# Make the four functions 

```python
import math

def f(x) : # f(x) = x^2
    return x*x

def f_log(x) : # f(x) = log(x)
    return math.log(x)

def f_sin(x) : # f(x) = sin(x)
    return math.sin(x)

def f_tan(x) : # f(x) = tan(x)
    return math.tan(x)
```

# Define the derivative

```python
def cdf(f,x,h): 
    
    df = (f(x+h) - f(x-h)) / (2*h)
    
    return df
```

# Creating the domains for four functions

```python
import numpy as np
x = np.linspace(-10, 10, 500) ## list for f(x)=x^2, with domain [-10,10] 
                              ## f_3(x)=sin(x), with domain [-10,10]
                              ## list for f_4(x)=tan(x), with domain [-10,10]
        
x_log = np.linspace(0.5, 10, 500) ## list for f_2(x)=log(x), with domain [1*10^{-6},10]

x_tan = np.linspace(-0.4* np.pi, 0.4* np.pi, 500)  # Define the range, say -2π to 2π
                # Filter values where cos(x) is close to zero (e.g., within some small epsilon)
epsilon = 1e-5  # you can adjust this value based on precision needs
filtered_values = x_tan[np.abs(np.cos(x_tan)) > epsilon]
```

# Creating lists for f(x) values

```python
y = [] # create a new list y, for the image for function f(x)=x^2
y_log = [] # create a new list y, for the image for function f_2(x)=log(x)
y_sin = [] # create a new list y, for the image for function f_3(x)=sin(x)
y_tan = [] # create a new list y, for the image for function f_4(x)=tan(x)

for i in x : #use for loop, add image of the function into the list at the back
    y.append(f(i))
    y_sin.append(f_sin(i))
    
for j in x_log :
    y_log.append(f_log(j))
    
for k in x_tan :
    y_tan.append(f_tan(k))
```

# Creating lists for df(x) values

```python
yprime = [] # create a new list for the image of derivative under x^2
yprime_log = [] # create a new list for the image of derivative under log(x)
yprime_sin = [] # create a new list for the image of derivative under sin(x)
yprime_tan = [] # create a new list for the image of derivative under tan(x)

for i in x :
    yprime.append(cdf(f,i,1e-6))
    yprime_sin.append(cdf(f_sin,i,1e-6))
    
for j in x_log :
    yprime_log.append(cdf(f_log,j,1e-6))
    
for k in x_tan :
    yprime_tan.append(cdf(f_tan,k,1e-6))
```

# Creating tables for four functions 

```python
import pandas as pd
table1 = pd.DataFrame({"x": x, "y": y, "CDF Values":yprime})
table2 = pd.DataFrame({"x": x_log, "y": y_log, "CDF Values":yprime_log})
table3 = pd.DataFrame({"x": x, "y": y_sin, "CDF Values":yprime_sin})
table4 = pd.DataFrame({"x": x_tan, "y": y_tan, "CDF Values":yprime_tan})
```
